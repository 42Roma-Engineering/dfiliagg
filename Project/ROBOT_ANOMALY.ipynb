Inoltre secondo me sarebbe possibile utilizzare la distanza tra i centroidi come misura per indicare lo stato di funzionamento del sistema. In questo modo dovrei riuscire ad indicare lo stato del sistema dopo ogni ciclo di lavoro, evidenziando in anticipo se i macchinari stanno progressivmente perdendo prestazioni.

Adesso però proviamo ad addestrare un modello utilizzando l'algoritmo K-Means fuzzy. Ho scelto di utilizzare questo approccio perché mi permette di valutare la percentuale di appartenenza a ciascun cluster, così da migliorare ulteriormente le performance del modello.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import Normalizer

pca = PCA(n_components=3)
scaler = Normalizer()

def prep_data(directory, col):
    dfs = []

    for filename in os.listdir(directory):
        if filename.endswith(".csv"):
            df = pd.read_csv(os.path.join(directory, filename), names=col, sep=';')
            df = df.drop(to_drop, axis=1)

            df_stat = df.describe()
            df_stat = df_stat.iloc[1:, :6]
            df_stat = pd.DataFrame([df_stat.values.flatten()])
            df = pd.DataFrame([df.corr().iloc[:,:].values.flatten()])
            df = pd.concat([df, df_stat], axis=1).reset_index(drop=True)
            df = df.dropna(axis=1)
            dfs.append(df)

    merged_df = pd.concat(dfs, ignore_index=True)
    return merged_df

df_normal = prep_data("./rawData/nominal", col)
df_oil = prep_data("./rawData/oilLeakage", col)
df_dry = prep_data("./rawData/dryBearing", col)
df_payload = prep_data("./rawData/noPayload", col)

#concateno i dataframe riga per riga
df = pd.concat([df_normal, df_oil, df_dry, df_payload], axis=0, ignore_index=True)
df = scaler.fit_transform(df)
df = pca.fit_transform(df)
```

```python
import skfuzzy as fuzz

cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(df.T, 4, 2, error=0.005, maxiter=3000, init=None)

```

```python
def test(path):
    path= path+"/*.csv"
    test_files = glob.glob(path)
    i=0
    for file in test_files:
        i += 1
        test = pd.read_csv(file, names=col, sep=";")

        df_test = test.drop(to_drop, axis=1)
        df_test_stat = df_test.describe()
        df_test_stat = df_test_stat.iloc[1:, :6]
        df_test_stat = pd.DataFrame([df_test_stat.values.flatten()])
        df_test = pd.DataFrame([df_test.corr().iloc[:,:].values.flatten()])
        df_test = pd.concat([df_test, df_test_stat], axis=1).reset_index(drop=True)
        df_test = df_test.dropna(axis=1)
        df_test = scaler.transform(df_test)
        df_test = pca.transform(df_test)
        predicted_membership, _, _, _, _, _ = fuzz.cluster.cmeans_predict(
        df_test.T, cntr, 2, error=0.005, maxiter=3000, seed=42
        )

        cluster_predetto = np.argmax(predicted_membership[:, 0])
        print(f"CICLO {i}: STATO DI FUNZIONAMENTO {cluster_predetto + 1}")
    
        print("  - STATO MANUTENZIONE:")
        for j in range(4):
            print(f"    Cluster {j + 1}: {predicted_membership[j, 0] * 100:.2f}%")
        print("\n")
```
